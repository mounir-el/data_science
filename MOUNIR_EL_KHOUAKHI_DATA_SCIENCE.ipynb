```python
#recherche par code postale
#retourne les localisation des restaurant qui ont le même code postale saisi dans la variable code_postal, avec le nom des restaurant en popup

code_postal = "75001"
request = df[df['CP'].str.contains(code_postal)]

m = folium.Map(
    location=[48.85, 2.35],
    zoom_start=14,
    tiles='Stamen Terrain'
)

for i in request.index:
    lang = request['log'].loc[i]
    lat = request['lat'].loc[i]
    
    m.add_child(folium.LatLngPopup())

    folium.Marker(
        [lang, lat],
        popup=request['name'].loc[i]
    ).add_to(m)
    



m.add_child(folium.ClickForMarker(popup='Waypoint'))

m
```

```python
#recherche par spécialité
#retourne les localisation des restaurant de la même spécialité saisi dans la variable spec, avec le nom des restaurant en popup

spec = "Japonais"
request = df[df['spécialité'].str.contains(spec)]

m = folium.Map(
    location=[48.85, 2.35],
    zoom_start=14,
    tiles='Stamen Terrain'
)

for i in request.index:
    lang = request['log'].loc[i]
    lat = request['lat'].loc[i]
    
    m.add_child(folium.LatLngPopup())

    folium.Marker(
        [lang, lat],
        popup=request['name'].loc[i]
    ).add_to(m)
    



m.add_child(folium.ClickForMarker(popup='Waypoint'))

m
```
